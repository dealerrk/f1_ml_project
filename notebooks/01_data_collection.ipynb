# Formula 1 Race Strategy Intelligence System

## Notebook 1: Data Collection

Objective:
Collect lap data, telemetry, weather data, and session information from FastF1 for selected Formula 1 races.

```python
import pandas as pd
import numpy as np
import fastf1
```

```python
# loading the first race

session = fastf1.get_session(2024, "Bahrain", "R")
session.load()
session
```

```python
# getting the lap data

laps = session.laps
print(laps.shape)
laps.head()
```

```python
# exploring the columns

print(laps.columns.tolist())
```

```python
# keeping only the useful columns

lap_data = laps[
    [
        "Driver",
        "LapNumber",
        "LapTime",
        "Compound",
        "TyreLife",
        "Stint",
        "Position"
    ]
]
lap_data.head()
```

```python
# checking the null values or the missing values

print(lap_data.isnull().sum())
```

```python
# saving the dataset we have obtained

import os
os.makedirs("../data/raw", exist_ok=True)
lap_data.to_csv(
    "../data/raw/bahrain_2024_laps.csv",
    index=False
)
print("Dataset saved successfully!")
```
