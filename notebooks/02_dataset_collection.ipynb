# Notebook 2: Multi-Race Dataset Collection

## Objective

Automatically collect lap-by-lap race data from multiple Formula 1 races
across different seasons and save them as CSV files.


```python
import os
import fastf1
import pandas as pd
import numpy as np
```

```python
# enabling the cache

CACHE_DIR = "../data/f1_cache"
os.makedirs(CACHE_DIR, exist_ok=True)
fastf1.Cache.enable_cache(CACHE_DIR)
```

```python
# choosing our preferred races
# these will give good mix of high, low , medium-high, high+ street circuit degradation

races = [
    "Bahrain",
    "Spain",
    "Monza",
    "Singapore"
]
years = [2022, 2023, 2024]
```

```python
for year in years:
   for race in races:

    print(f"\nDownloading {year} {race}...")

    try:
        session = fastf1.get_session(year, race, "R")
        session.load()

        laps = session.laps

        important_columns = [
            "Driver",
            "Team",
            "LapNumber",
            "LapTime",
            "Compound",
            "TyreLife",
            "Stint",
            "Position",
            "TrackStatus"
        ]

        lap_data = laps[important_columns].copy()

        # Add Year and Race columns
        lap_data["Year"] = year
        lap_data["Race"] = race

        # Reorder columns
        lap_data = lap_data[
            [
                "Year",
                "Race",
                "Driver",
                "Team",
                "LapNumber",
                "LapTime",
                "Compound",
                "TyreLife",
                "Stint",
                "Position",
                "TrackStatus"
            ]
        ]

        filename = f"../data/raw/{year}_{race}_laps.csv"
        lap_data.to_csv(filename, index=False)

        print(f"Saved: {filename}")

    except Exception as e:
        print(f"Failed: {race}")
        print(e)
      
```
