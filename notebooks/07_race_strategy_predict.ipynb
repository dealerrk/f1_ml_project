## Notebook 7: Race Strategy Prediction System
Use the trained Random Forest model to predict lap times for new race scenarios and provide simple strategy insights.

```python
import pandas as pd
import numpy as np
import joblib
```

```python
# loading the trained model

model = joblib.load("../models/best_model.pkl")
```

```python
# creating the sample race scenario (instead of using the training data , creating new examples)

sample_data = pd.DataFrame({
    "LapNumber": [10, 25, 50],
    "TyreLife": [1, 14, 24],
    "Stint": [1, 2, 3],
    "Position": [1, 3, 5],
    "TrackStatus": [1, 1, 1],
    "FuelLoadApprox": [0.80, 0.55, 0.30],
    "StintProgress": [0.20, 0.60, 0.95],
    "CompoundEncoded": [0, 1, 2],
    "FreshTyre": [1, 0, 0],
    "TyreAgeSquared": [1, 196, 576]
})

sample_data
```

```python
# predicting the lap times

sample_data["PredictedLapTime"] = model.predict(sample_data)
sample_data
```

```python
# adding a simple strategy recommendation

def recommend_pitstop(row):
    if row["StintProgress"] > 0.85:
        return "Pit Recommended"

    if row["FreshTyre"] == 1:
        return "Stay Out"

    return "Monitor Tyres"

sample_data["Recommendation"] = sample_data.apply(
    recommend_pitstop,
    axis=1
)

sample_data
```

```python
sample_data.to_csv(
    "../outputs/tables/strategy_predictions.csv",
    index=False
)

sample_data
```
